## Creating Random tensors

Randoms tensors are tensors of some arbitrary size which contain random numbers

```python
# Create two random (but the same) tensors

random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42)
random_1 = random_2.normal(shape=(3,2))

```

## Shuffle the order of elements in a tensor

```python
# Shuffle a tensor (valuable for when you want to shuffle your data so the inherit order does not affect learning )
not_shuffle = tf.constant([[10,7],
                           [3,4],
                           [2,5]])
# Shuffle our non-shuffle tensor
tf.random.set_seed(42)
tf.random.shuffle(not_shuffle,seed=42)
```

```python
tf.random.set_seed(42) #global level random seed
tf.random.shuffle(not_shuffle,seed=42) #operation level random seed
```

## Getting more information from tensors
When dealing with tensors you probablu want to be aware of the folowing attributes
* Shape
* Rank
* Axis or dimension
* Size

```python
# Create a rank 4 tensor ( 4 dimensions)
rank_4_tensor = tf.zeros(shape=[2,3,4,5])
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape,rank_4_tensor.ndim,tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor
print("datatupe of every element:", rank_4_tensor.dtype)
print("number of dimensions (rank):", rank_4_tensor.ndim)
print("shape of tensor:", rank_4_tensor.shape)
print("elements along the 0 axis:", rank_4_tensor.shape[0])
print("elements along the last axis:", rank_4_tensor.shape[-1])
print("total number of elements in our tensor:", tf.size(rank_4_tensor).numpy())
```

## Indexing tensor
Tensors can be indexing just like Python lists

```python
# Get the firts 2 elements of each dimension
rank_4_tensor[:2,:2,:2,:2]
```

```python
# Get the first element from each dimension from each index excep for the final one
rank_4_tensor[:1,:1,:1,:]
```

```python
# Create a rank 2 tensor
rank_2_tensor = tf.zeros(shape=[2,2])
rank_2_tensor
```

```python
# Get the last item of each our rank 2 tensor
rank_2_tensor[:,-1]
```

```python
# Add in extra dimension to our rank 2 tensor
rank_3_tensor = rank_2_tensor[...,tf.newaxis]
rank_3_tensor
```

```python
# Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor,axis=-1)
```

```python
# Expand the axis 0
tf.expand_dims(rank_2_tensor,axis=0)
```

## Changing the datatype of the tensor

```python
# Create a new tensor with default datatype (float32)
B = tf.constant([1.7,7.4])
B.dtype
```

```python
C = tf.constant([7,10])
C.dtype
```

```python
# Change from float32 to float16 (reduced precision)
D = tf.cast(B,dtype=tf.float16)
D , D.dtype
```

```python
# Change from int32 to float32
E = tf.cast(C,dtype=tf.float32)
E,E.dtype
```

```python
E_float16 = tf.cast(E,dtype=tf.float16)
E_float16
```

## Aggreagation tensors

Aggregating tensor = condensing them from multiples values down to a smaller amount of values.

```python
# Get the absolute values
D = tf.constant([-7,-10])
tf.abs(D)
```

### Let's go through the following forms of aggregation:
* Get the min
* Get the max
* Get mean of a tensor
* Get the sum of a tensor

```python
# Create a random tensor with values between 0 and 100 of size 50
import numpy as np
E = tf.constant(np.random.randint(0,100,size=50))
E
```

```python
tf.size(E), E.shape,E.ndim
```

```python
# Find the min of a tensor
tf.reduce_min(E)
```

```python
# Find the max of a tensor
tf.reduce_max(E)
```

```python
# Find the mean
tf.reduce_mean(E)
```

```python
# Find the sum
tf.reduce_sum(E)
```

```python
# Find the variance
import tensorflow_probability as tfp
tfp.stats.variance(E)
```

```python
# Find the standard desviation
tf.math.reduce_std(tf.cast(E,dtype=tf.float32))
```

## Find the positional maximun and minimun

```python
# Create a new tensor for finding positional minimun and maximun
tf.random.set_seed(42)
F =tf.random.uniform(shape=[50])
F
```

```python
# Finding the positional maximun
tf.argmax(F)
```

```python
# Index on our largest value position
F[tf.argmax(F)]
```

```python
# Finding the max of F
tf.reduce_max(F)
```

```python
# Check for equality
F[tf.argmax(F)] == tf.reduce_max(F)
```

```python
tf.argmin(F)
```

```python
F[tf.argmin(F)]
```

```python
tf.reduce_min(F)
```
